## INFO 623 - Social Network Analytics

```python
import requests, re
from bs4 import BeautifulSoup

marvel_url = "https://www.imdb.com/title/tt4154796/fullcredits?ref_=tt_ql_1"

mcu = requests.get(url=marvel_url)
soup = BeautifulSoup(mcu.text, 'html.parser')

for td in soup.body.find_all('td'):
    if td.get('class') is not None:
        # Vertified cast in credits order
        if td['class'][0] == 'character':
            for element in td.find_all(re.compile('a')):
                print(element.text)
        # Uncredited characters 
        if td.text == 'Rest of cast listed alphabetically:':
            break
```
